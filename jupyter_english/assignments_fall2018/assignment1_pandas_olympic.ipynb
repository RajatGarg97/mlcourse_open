__1. How old were the youngest male and female participants of the 1996 Olympics?__

- 16 and 15
- 14 and 12 
- 16 and 12
- 13 and 11

```python
# You code here 
df = data.loc[data['Year'] == 1996]
df_sort = df.sort_values(by=['Age'])
df_M = df_sort.loc[df_sort['Sex']=='M']
df_F = df_sort.loc[df_sort['Sex']=='F']

df_M
df_F
# Answer: 14 and 12
```

__3. What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? Round the answer to the first decimal.__

- 178.5 and 7.2
- 179.4 and 10
- 180.7 and 6.7
- 182.4 and 9.1 

```python
# You code here 

df_F = df_Y.loc[df_Y['Sex'] == 'F']
df_bb = df_F.loc[df_F['Sport'] == 'Basketball']
print(df_bb['Height'].mean())
print(df_bb['Height'].std())
# Answer: 182.4 and 9.1
```

__7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 

- Yes 
- No

```python
# You code here 
df_Y3 = data.loc[data['Year'] == 2016]
df_Y3Sw = df_Y3.loc[df_Y3['Team'] == 'Switzerland']
df_Y3Se = df_Y3.loc[df_Y3['Team'] == 'Siberia']
print(df_Y3Sw['Medal'].value_counts().sum())
print(df_Y3Se['Medal'].value_counts().sum())
# Answer: No
```

__9. Is it true that there were Summer Olympics held in Lake Placid? Is it true that there were Winter Olympics held in Sankt Moritz?__

- Yes, Yes
- Yes, No
- No, Yes

__10. What is the absolute difference between the number of unique sports at the 1995 Olympics and 2016 Olympics?__

- 16
- 24
- 26
- 34

```python
# You code here 
df_Y95 = data.loc[data['Year'] == 1995]
df_Y16 = data.loc[data['Year'] == 2016]
print(df_Y16.drop_duplicates(subset='Sport', keep='first', inplace=False)['Sport'].count())
print(df_Y95.drop_duplicates(subset='Sport', keep='first', inplace=False)['Sport'].count())
# Answer: 34
```
